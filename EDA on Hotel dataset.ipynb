# DATA CLEANING

```python
df.drop(columns = ['company', 'agent'], inplace = True)
```

```python
df = df[df['adults'] != 0]
```

```python
df = df[df['adr'] > 0]
```

```python
df['arrival_date'] = df['arrival_date_month'] + " " + df['arrival_date_day_of_month'].astype(str) + " " + df['arrival_date_year'].astype(str)
df['arrival_date'] = pd.to_datetime(df['arrival_date'])
df['arrival_date']
```

```python
df.drop('arrival_date_week_number', inplace = True, axis = 1)
```

```python
df['country'].mode()
```

```python
df.loc[df['country'] == 'PRT', 'country'].count()
```

```python
df['country'].fillna(df['country'].mode()[0], inplace = True)
```

```python
df.isnull().sum()
```

```python
df['children'].median()
```

```python
df['children'].fillna(df['children'].median(), inplace = True)
```

```python
df.isnull().sum()
```

```python
df['total_guests'] = (df['adults'] + df['children'] + df['babies']).astype(int)
```

```python
graph = sns.countplot(data = df, x = 'hotel')
graph.set(title = 'Hotel type and Reservation Count')
graph.set_xlabel('Hotel Type')
graph.set_ylabel('Reservation_count')
```

```python
df['hotel'].value_counts(normalize = True)
```

```python
graph = sns.countplot(data = df, x = 'hotel', hue = 'is_canceled')
graph.set(title = 'Hotel type and No. of Cancellations/ non-cancellations')
graph.set_xlabel('Hotel Type')
graph.set_ylabel('No. of Cancellations/ non-cancellations')
```

```python
plt.figure(figsize = (12, 12))
graph = sns.countplot(data = df,x = 'arrival_date_month',
              hue = 'hotel', 
              order = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"])
graph.set(title = 'Hotel type and NO. of arrivals per Month')
graph.set_xlabel('Hotel Type')
graph.set_ylabel('NO. of arrivals per Month')
```

```python
graph = sns.countplot(data = df, x = 'adults', hue = 'hotel')
graph.set(title = 'Number of adults reserving')
graph.set_xlabel('No. of Adults')
graph.set_ylabel('Reservation Count')
```

```python
graph = sns.countplot(data = df, x = 'children', hue = 'hotel')
graph.set(title = 'Number of families with children reserving')
graph.set_xlabel('No. of Children')
graph.set_ylabel('Reservation Count')
```

```python
sns.set(rc = {'figure.figsize' : (14, 12)})
fig, ax = plt.subplots()
sns.lineplot(
    df[df['hotel'] == 'City Hotel'].groupby(pd.Grouper(key = 'arrival_date', axis = 0,
                                                                freq = 'M')).sum().reset_index(),
                x = 'arrival_date',
                y = 'total_guests',
                ax = ax
)
sns.lineplot(
    df[df['hotel'] == 'Resort Hotel'].groupby(pd.Grouper(key = 'arrival_date', axis = 0,
                                                                freq = 'M')).sum().reset_index(),
                x = 'arrival_date',
                y = 'total_guests',
                ax = ax
    
)
ax.set(title = 'Number of Guests each Month by Hotel Type')
ax.set_xlabel('Month-Year')
ax.set_ylabel('Total Guest Count')
```

```python
sns.set(rc = {'figure.figsize' : (14, 12)})
fig, ax = plt.subplots()
sns.lineplot(
    df[df['hotel'] == 'City Hotel'].groupby(pd.Grouper(key = 'arrival_date', axis = 0,
                                                                freq = 'M')).sum().reset_index(),
                x = 'arrival_date',
                y = 'required_car_parking_spaces',
                ax = ax
)
sns.lineplot(
    df[df['hotel'] == 'Resort Hotel'].groupby(pd.Grouper(key = 'arrival_date', axis = 0,
                                                                freq = 'M')).sum().reset_index(),
                x = 'arrival_date',
                y = 'required_car_parking_spaces',
                ax = ax
    
)
ax.set(title = 'Monthly required Parking Spaces')
ax.set_xlabel('Month-Year')
ax.set_ylabel('Parking Spaces Requirwd')
```
